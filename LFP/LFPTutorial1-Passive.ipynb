# LFP Tutorial
This is an explanation of the calculation of the extracellular potential. To run through this example code and produce output, press **``<shift-Enter>``** in each code block below.

First step is to **import ``LFPy``** and other packages for analysis and plotting:

```python
import LFPy
```

Create some dictionarys with parameters for cell, synapse and extracellular electrode:

```python
cellParameters = {
    'morphology' : 'single_section.hoc',
    'tstart' : -50,
    'tstop' : 100,
    'dt' : 2**-4,
    'passive': True,
    'passive_parameters': {'g_pas':0.001, 'e_pas':-70},
    'nsegs_method': 'fixed_length',
    'max_nsegs_length':6
}

SynapseParameters = {
    'syntype' : 'Exp2Syn',
    'e' : -20,
    'tau1' : 0.5,
    'tau2' : 2.0,
    'weight' : 0.009,
    'record_current' : True,
}

z = mgrid[-400:1201:100]
electrodeParameters = {
    'x' : zeros(z.size),
    'y' : zeros(z.size),
    'z' : z,
    'sigma' : 0.3,
}
```

Then, create the **cell**, **synapse** and **electrode** objects using 
**LFPy.Cell**, **LFPy.Synapse**, **LFPy.RecExtElectrode**.

```python
# Cell 1
cell = LFPy.Cell(**cellParameters)
cell.set_pos(x=-250, y=0, z=0)
cell.set_rotation(z=np.pi)


synapse = LFPy.Synapse(cell,
                       idx = cell.get_closest_idx(x=-250,z=100),
                       **SynapseParameters)

synapse.set_spike_times(array([50]))       

electrode = LFPy.RecExtElectrode(**electrodeParameters)
```

Run the simulation with **electrode** as input to **cell.simulate()**

```python
cell.simulate(electrode=electrode,rec_imem=True,rec_ipas=True,rec_icap=True)
```

Then plot the **somatic potential** and the **LFP**
from electrode object:

```python
figure(figsize=(15, 10))
#subplot(2,3,4)


subplot(1,3,1)
# Plot geometry of the cell
for sec in LFPy.cell.neuron.h.allsec():
    idx = cell.get_idx(sec.name())
    plot(np.r_[cell.xstart[idx], cell.xend[idx][-1]],
            np.r_[cell.zstart[idx], cell.zend[idx][-1]],
            'k',linewidth=8)
# Plot synapse as red dot
plot([cell.synapses[0].x], [cell.synapses[0].z], \
     color='r', marker='o', markersize=5,label='synapse')
# Plot electrodes as green dots
plot(electrode.x, electrode.z, '.', marker='o', color='g',label='electrodes')
x = np.arange(-250,0)
yr = 100*np.ones((250,))
plot(x,yr,color=[0.5,0.5,0.5]),text(-125,95,'r')
h = -250*np.ones((86,))
hy = np.arange(14,100)
plot(h,hy,color=[0.5,0.5,0.5]),text(-240,50,'h')
l = -200*np.ones((113,))
ly = np.arange(-12.5,100)
plot(l,ly,color=[0.5,0.5,0.5]), text(-180,40,'l')
xlabel('distance (um)')
ylabel('distance (um)')

ylim(-150,120)
legend()

sp = [14,11,8,5,2]
segs = [5,4,3,2,1]
for i in np.arange(1,6):
    subplot(5,3,sp[i-1])
    plot(cell.tvec,cell.imem[i-1,:],'blue',label='imembrane')
    plot(cell.tvec,cell.ipas[i-1,:],'red',label='i_pas')
    plot(cell.tvec,cell.icap[i-1,:],'green',label='i_cap')
    xlim(42.5,62.5)
    if i!=5:
        text(45,0.06,'segment {}'.format(segs[i-1]))
    else:
        text(45,-0.1,'segment {}'.format(segs[i-1]))
        plot(cell.tvec, synapse.i,'m',label='i_syn')
        legend(loc='lower right')
        title('currents (nA)')
    if i==1:
        xlabel('time (ms)')
        
#### Calculate extracellular potential by hand

# electrode 5 is at 0,100
h = np.zeros((5,1))
l = np.zeros((5,1))
r = np.zeros((5,1))

for i in np.arange(0,5):
    h[i,0] = electrode.z[5]-cell.zend[i]
    l[i,0] = electrode.z[5]-cell.zstart[i]
    r[i,0] = electrode.x[5]-cell.xstart[i]

dist = np.log((np.sqrt(h**2+r**2)-h)/(np.sqrt(l**2+r**2)-l))
v_ext=1000*(1/(4*np.pi*300*5))*np.dot(np.transpose(cell.imem),dist)

for i in np.arange(1,4):
    ax=subplot(3,3,3*i)
    plot(cell.tvec,electrode.LFP[5-(i-1),:],'k')
    xlim(42.5,62.5)
    text(45,0.000002,'electrode {}'.format(i))
    ax.yaxis.tick_right()
    if i==1:
        plot(cell.tvec,v_ext,'k')
        title('extracellular voltage (mV)')
    if i==3:
        xlabel('time(ms)')
    ylim(-0.0000055,0.0000055)
```
